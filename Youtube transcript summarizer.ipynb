# **Audio Chunking**

It is used to avoid out of memory and is used to clipping or splitting the audio into multiple smaller chunks and writing them into individual files

```python
import librosa
```

```python
input_file = '/content/ytaudio.wav'
```

```python
print(librosa.get_samplerate(input_file))
stream = librosa.stream(
    input_file,
    block_length = 30,
    frame_length = 16000,
    hop_length = 16000
)
```

```python
import soundfile as sf
```

```python
for i, speech in enumerate(stream):
  sf.write(f'{i}.wav', speech, 16000)
```
